```python
df = df[~df.description.isnull()]
df['description_length'] = df.description.apply(lambda x: len(x))
max(df.description_length)
```

```python
descriptions = df.description.tolist()
```

```python
import numpy as np
from transformers import AutoTokenizer, AutoModel
import torch
from sklearn.metrics.pairwise import cosine_similarity
```

```python
model_name = 'dccuchile/bert-base-spanish-wwm-uncased'
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
```

```python
def prepare_data_for_bert(texts, tokenizer, max_length):
    input_ids = []
    attention_masks = []
    encodeds_data = []

    for text in texts:
        encoded_data = tokenizer.encode_plus(
            text,
            add_special_tokens=True,
            max_length=max_length,
            pad_to_max_length=True,
            return_attention_mask=True,
            return_tensors='pt'
        )
        input_ids.append(encoded_data['input_ids'])
        attention_masks.append(encoded_data['attention_mask'])
        encodeds_data.append(encoded_data)

    input_ids = torch.cat(input_ids, dim=0)
    attention_masks = torch.cat(attention_masks, dim=0)

    return input_ids, attention_masks, encodeds_data
```

```python
input_ids, attention_masks, encodeds_data = prepare_data_for_bert(descriptions, tokenizer, 1342)
```

```python
df['tokens'] = [' '.join(tokenizer.convert_ids_to_tokens(encodeds_data[i]['input_ids'][0])) for i in range(len(encodeds_data))]
```

```python
df['tokens'] = df['tokens'].str.replace(' #', '')
df['tokens'] = df['tokens'].str.replace('#', "")
df['tokens'] = df['tokens'].str.replace('\[CLS\]', "")
df['tokens'] = df['tokens'].str.replace('\[PAD\]', "")
df['tokens'] = df['tokens'].str.replace('\[SEP\]', "")
df['tokens'] = df['tokens'].str.replace('\[UNK\]', "")
df['tokens'] = df['tokens'].str.replace('\[MASK\]', "")
df['tokens'] = df['tokens'].str.strip()
```

```python
a = df.tokens.tolist()
for token in a:
  print(token)
```

```python
df.to_csv('/content/drive/MyDrive/peru_larco/Data/base_n150/base_n150.csv', sep=';', index=False)
```

# 4. Eliminamos stopwords y creamos la variable de número de palabras

```python
from google.colab import drive
```

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/peru_larco/Data/base_clean/base_global.csv',  sep=';')
```

Eliminamos un dato que está missing

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
import nltk

nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
```

```python
stop_words = set(stopwords.words("spanish"))
lemmatizer = WordNetLemmatizer()
```

```python
def tokenize(description, lemmatizer, stop_words) -> str:
    return ' '.join([
    lemmatizer.lemmatize(word.lower())
        for word in word_tokenize(description) if word.isalnum() and word.lower() not in stop_words
    ])
```

```python
df['tokens'] = df.apply(lambda row: tokenize(row['tokens'], lemmatizer, stop_words), axis=1)
```

```python
for indice, fila in df.iterrows():
    print("Descripción:", fila['description'])
    print("Tokens:", fila['tokens'])
    print()
```

Aplicamos un contador de palabras

```python
def contar_palabras(texto):
    palabras = texto.split()
    num_palabras = len(palabras)
    return num_palabras
```

```python
df['numero_palabras'] = df['tokens'].apply(contar_palabras)
```

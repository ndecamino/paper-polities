# Generamos la matriz

```python
distances= dict()
cultures = list()
with open(path, 'r') as file:
    for line in file:
        _, pair, _, m_distance = line.strip().split(' ')
        c1, c2 = pair.split('-')
        if c1 not in cultures:
            cultures.append(c1)
        if c2 not in cultures:
            cultures.append(c2)

        distances[(c1, c2)] = np.float64(m_distance)

distances
```

```python
df = pd.DataFrame(index=sorted(cultures), columns=sorted(cultures))

for c in cultures:
    df[c] = df[c].astype(np.float64)
df.head()
```

```python
for c1 in cultures:
    for c2 in cultures:
        if c1 == c2:
            df[c1][c2] = 0
        elif (c1, c2) in distances.keys():
            df[c1][c2] = distances[(c1, c2)]
            df[c2][c1] = distances[(c1, c2)]
        elif (c2, c1) in distances.keys():
            df[c1][c2] = distances[(c2, c1)]
            df[c2][c1] = distances[(c2, c1)]
        else:
            df[c1][c2] = 0.0
            df[c2][c1] = 0.0
```

```python
df
```

```python
max_mean_distance = np.max(list(distances.values()))
min_mean_distance = np.min(list(distances.values()))
std_mean_distance = np.std(list(distances.values()))
```

```python
fig, ax = plt.subplots(figsize=(15, 12))
plt.title(path)
heatmap = sns.heatmap(df,
            vmax=max_mean_distance+std_mean_distance,
            vmin=min_mean_distance-std_mean_distance,
            cmap='coolwarm',
            linewidths=.5,
            square=True,
            annot=True,
            fmt=".4f",
            mask=np.triu(df),
            ax=ax)
plt.show()
```

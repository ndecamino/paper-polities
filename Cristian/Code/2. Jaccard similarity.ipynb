# 1. Definimos Jaccard similarity

Primero creamos funciones auxiliares

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import math
import time
```

```python
def generate_shinglings(text, k):
    shinglings = set()

    if len(text) < k:
        raise ValueError("El tamaño del texto debe ser mayor o igual al valor de k.")


    for i in range(len(text) - k + 1):
        shingling = text[i:i+k]
        shinglings.add(shingling)

    return shinglings
```

```python
def auxiliar(set1, set2):

    intersection_size = len(set1.intersection(set2))
    union_size = len(set1.union(set2))

    # Calcular la similitud de Jaccard
    if union_size == 0:
        return 0.0
    else:
        return intersection_size / union_size
```

Ahora creamos las dos funciones de jaccard similarity: shinglings y tokens

```python
def jaccard_similarity(string1,string2,k):
  if len(string1) < k or len(string2) < k:
    return 0
  set1 = generate_shinglings(string1,k)
  set2 = generate_shinglings(string2,k)

  similarity = auxiliar(set1,set2)
  return similarity
```

```python
def jaccard_similarity_tokens(str1, str2):
    # Convierte cada string en un conjunto de palabras
    set1 = set(str1.split())
    set2 = set(str2.split())

    # Calcula la intersección y la unión de los conjuntos
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))

    # Calcula la similitud de Jaccard
    similarity = intersection / union if union != 0 else 0

    return similarity
```

Ejemplos de uso

```python
texto_ejemplo = "abcdefgh"
k_elegido = 3
shinglings_resultantes = generate_shinglings(texto_ejemplo, k_elegido)
print(shinglings_resultantes)
```
